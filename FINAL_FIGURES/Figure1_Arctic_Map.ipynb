```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import cartopy.crs as ccrs
from glob import glob
import numpy as np
import math

from matplotlib.pyplot import figure
import pylab 
```

```python
file = '/glade/derecho/scratch/gabyn/SPECTRA_hindcast/growth_lim'
diat3_val = xr.open_mfdataset(f'{file}/diat3_lim_value_remap.nc', decode_times=False)
```

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(1, 1, 1)
m = Basemap(projection='npstere',round=True,
            boundinglat=65,
            lon_0=180+(-100+-30)/2.,resolution='c')

m.drawmeridians((180,205), color='k', textcolor='k', linewidth=2.0, fontsize=15,
              zorder=None, dashes=[1, 1], labels=[1, 1, 1, 1])
m.drawmeridians((205,235), color='k', textcolor='k', linewidth=2.0, fontsize=15,
              zorder=None, dashes=[1, 1], labels=[1, 1, 1, 1])
m.drawmeridians((235,260), color='k', textcolor='k', linewidth=2.0, fontsize=15,
              zorder=None, dashes=[1, 1], labels=[1, 1, 1, 1])
m.drawmeridians((260,315), color='k', textcolor='k', linewidth=2.0, fontsize=15,
              zorder=None, dashes=[1, 1], labels=[1, 1, 1, 1])
m.drawmeridians((145,180), color='k', textcolor='k', linewidth=2.0, fontsize=15,
              zorder=None, dashes=[1, 1], labels=[1, 1, 1, 1])
m.drawmeridians((90,145), color='k', textcolor='k', linewidth=2.0, fontsize=15,
              zorder=None, dashes=[1, 1], labels=[1, 1, 1, 1])
m.drawmeridians((53,90), color='k', textcolor='k', linewidth=2.0, fontsize=15,
              zorder=None, dashes=[1, 1], labels=[1, 1, 1, 1])
m.drawmeridians((20,53), color='k', textcolor='k', linewidth=2.0, fontsize=15,
              zorder=None, dashes=[1, 1], labels=[1, 1, 1, 1])
m.drawparallels((60,90), color='k', textcolor='k', 
                linewidth=2.0, zorder=None, dashes=[1, 1], 
                labels=[0, 0, 0, 0])

x1, y1 = m(*np.meshgrid(diat3_val.lon[348],diat3_val.lat[158]))
m.plot(x1,y1,'X',color='k',ms=13,fillstyle='full')

#x1, y1 = m(*np.meshgrid(diat3_val.lon[2],diat3_val.lat[156]))
#m.plot(x1,y1,'o',color='k',ms=13,fillstyle='full')

x1, y1 = m(*np.meshgrid(diat3_val.lon[200],diat3_val.lat[175]))
m.plot(x1,y1,'*',color='k',ms=16,fillstyle='full')

x1, y1 = m(*np.meshgrid(diat3_val.lon[192],diat3_val.lat[158]))
m.plot(x1,y1,'^',color='k',ms=13,fillstyle='full')

m.drawcoastlines(linewidth=0.5, linestyle='solid', color='gray', antialiased=1, ax=None, zorder=None)
m.fillcontinents(color='0.8', lake_color=None, ax=None, zorder=None, alpha=None)
m.plot
plt.show()

fig.savefig('/glade/work/gabyn/case_notes/analysis_figures/Arctic_regions.png', dpi=600, bbox_inches='tight')
```
