```python
## Volume 
# units [um^3]
pp_v = 0.195428825646897
mp1_v = 37.6111075368976
diaz_v = 122.786536059293
diat1_v = 4187.92236603257
mp2_v = 7238.41739042113
diat2_v = 374617.983936749
mp3_v = 1393064.17037973
diat3_v = 33510323.645704
mp4_v = 268101116.324664

phyto_vol = array([pp_v,mp1_v, diaz_v, diat1_v,mp2_v,diat2_v,mp3_v,diat3_v,mp4_v])
```

## import abundance
    abundance = sio.loadmat('../Notebooks_Question2/mat_files/plankton_abundance_zavg_z.mat')
    phy_abundance = abundance['phy_ab']

## Calcualte the slopes for every phytoplankton and zooplankton type at every month. 
We do this, by using the "stats.linregress" function, this allows us to calculate: 
   * slope
   * y-intercept
   * r-values
   * p_values
   * standard errors. 
    
    We must mask out any NaNs that migth be in the data set
    as it turns out, the diazotrophs show no abundance and therefore we need to mask them, and since it is the same for every month we will just be using the same mask for the first month "mask = ~np.isnan(xp) & ~np.isnan(yp[:,0])" which is the same thing as "mask = [~np.isnan(xp) & ~np.isnan(yp[:,i]) for i in range(12)]"

# log of volumes for phytoplankton, zooplankton and both. 
xp = np.log10(phyto_vol+1)

# Make the zero's into NaNs 
phy_abundance[phy_abundance<0] = 0

# abundance for phyto
yp_surface = np.log10(phy_abundance[:,:,1,:,:]+1)
yp_100m    = np.log10(phy_abundance[:,:,10,:,:]+1)

# the size of 777600
n = 12*180*360

# reshape the 4-D array to a 2-D array. 
yp_reshape_surface = yp_surface.reshape((9, n))
yp_reshape_100m = yp_100m.reshape((9, n))

# create the calculation 
yp_results_surface = np.empty((n, 5))
yp_results_100m = np.empty((n, 5))

for i in range(n):
    yp_results_surface[i] = stats.linregress(xp, yp_reshape_surface[:,i])
    yp_results_100m[i] = stats.linregress(xp, yp_reshape_100m[:,i])
    
# reshape back to size wanted. back to 4D, but the 5 is where the slope, interecpt, etc. are. 
ypp_surface = yp_results_surface.reshape((12,180,360, 5)) 
ypp_100m = yp_results_100m.reshape((12,180,360, 5)) 

```python
import scipy.io as sio
sss = sio.loadmat('../Question_2/mat_files/size_spectral_slope_surface.mat')
ypp_surface = sss['phyto_regression_surface']
ypp_surface[ypp_surface==0]=np.nan
```

## try one with global average and one with full Arctic for simplicity

```python
slopes_arctic = np.nanmean(ypp_surface[:,149:180,:,0], axis=(1,2))
slopes_globe = np.nanmean(ypp_surface[:,:,:,0], axis = (1,2))
```

```python
import matplotlib
cmap=matplotlib.cm.get_cmap('Spectral', 10)

months = ('jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec')

fig = figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')
pylab.plot(months,slopes_arctic, '-o', color=cmap(0),label='Arctic Ocean')
pylab.plot(months,slopes_globe, '-o',color=cmap(1),label='Global Ocean')

pylab.legend(loc='lower right')
pylab.ylim([-1.1,-0.6])
pylab.ylabel('size-spectral slope', fontsize=16)
pylab.title('surface climatology', fontsize=16)
pylab.show()
```

## Artic Regions

```python
slopes_p_ao = np.nanmean(ypp_surface[:,170:180,:,0], axis=(1,2))
slopes_p_cs = np.nanmean(ypp_surface[:,149:170,0:25,0], axis = (1,2))
slopes_p_bs = np.nanmean(ypp_surface[:,149:170,25:55,0], axis = (1,2))
slopes_p_ca = np.nanmean(ypp_surface[:,149:170,55:110,0], axis = (1,2))
slopes_p_bb = np.nanmean(ypp_surface[:,149:170,110:135,0], axis = (1,2))
slopes_p_ess = np.nanmean(ypp_surface[:,149:170,325:360,0], axis = (1,2))
slopes_p_ls = np.nanmean(ypp_surface[:,149:170,270:325,0], axis = (1,2))
slopes_p_ns = np.nanmean(ypp_surface[:,149:170,155:205,0], axis = (1,2))
slopes_p_bts =np.nanmean(ypp_surface[:,149:170,205:233,0], axis = (1,2))
slopes_p_ks = np.nanmean(ypp_surface[:,149:170,233:270,0], axis = (1,2))
```

## avergae nitrate throughout the Arctic Ocean 

```python
no3_arctic = np.nanmean(ds.NO3[:,0,149:180,:], axis =(1,2))
no3_ns = np.nanmean(ds.NO3[:,0,149:170,155:205], axis =(1,2))
```

## Slopes log(Abundance) vs log(Volume)
### Units 
The units for the slope: 
    $\frac{rise}{run} = \frac{log(abundance)}{log(volume)} = \frac{\frac{ind}{L}}{\mu m^3} = \frac{\mu m^3}{L}$

## Plot just the slopes of the phytoplankton 

```python
import matplotlib
cmap=matplotlib.cm.get_cmap('Spectral', 10)

months = ('jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec')

fig = figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')
pylab.plot(months,slopes_p_ao, '-o', color=cmap(0),label='Central Arctic')
pylab.plot(months,slopes_p_cs, '-o',color=cmap(1),label='Chukchi Sea')
pylab.plot(months,slopes_p_bs, '-o', color=cmap(2),label='Beaufort Sea')
pylab.plot(months,slopes_p_ca, '-o', color=cmap(3),label='Canadian Archipelago')
pylab.plot(months,slopes_p_bb, '-o', color=cmap(4),label='Baffin Sea')
pylab.plot(months,slopes_p_ns, '-o', color=cmap(5),label='Nordic Sea')
pylab.plot(months,slopes_p_bts, '-o',color=cmap(6),label='Barents Sea')
pylab.plot(months,slopes_p_ks, '-o', color=cmap(7),label='Kara Sea')
pylab.plot(months,slopes_p_ls, '-o', color=cmap(8),label='Laptev Sea')
pylab.plot(months,slopes_p_ess, '-o', color=cmap(9),label='Eastern Siberian Sea')

pylab.legend(loc='lower right')
pylab.ylim([-1.1,-0.6])
pylab.ylabel('size-spectral slope', fontsize=16)
pylab.title('surface climatology', fontsize=16)
pylab.show()

fig.savefig('slopes_p_surface.png', dpi=300, bbox_inches='tight')
```
