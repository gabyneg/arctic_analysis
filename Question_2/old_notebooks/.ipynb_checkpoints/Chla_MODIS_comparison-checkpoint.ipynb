```python
%matplotlib inline
import warnings
warnings.simplefilter("ignore") # Silence warnings
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
import hvplot.xarray
import xarray as xr
import cartopy
import cartopy.crs as ccrs
import numpy as np
from matplotlib.pyplot import figure
import pylab 
from cartopy.util import add_cyclic_point

# Scientific libraries
from numpy import arange,array,ones
from scipy import stats

import scipy.io as sio
import pandas as pd
import matplotlib.cm as cm
import glob
import matplotlib.colors as colors
```

```python
ds7 = xr.open_dataset('modis/PAR_Climatology_Jul_remapgrid.nc', decode_times=False)
ds8 = xr.open_dataset('modis/PAR_Climatology_Aug_remapgrid.nc', decode_times=False)
ds9 = xr.open_dataset('modis/PAR_Climatology_Sep_remapgrid.nc', decode_times=False)
ds10 = xr.open_dataset('modis/PAR_Climatology_Oct_remapgrid.nc', decode_times=False)
ds11 = xr.open_dataset('modis/PAR_Climatology_Nov_remapgrid.nc', decode_times=False)
ds12 = xr.open_dataset('modis/PAR_Climatology_Dec_remapgrid.nc', decode_times=False)
ds1 = xr.open_dataset('modis/PAR_Climatology_Jan_remapgrid.nc', decode_times=False)
ds2 = xr.open_dataset('modis/PAR_Climatology_Feb_remapgrid.nc', decode_times=False)
ds3 = xr.open_dataset('modis/PAR_Climatology_Mar_remapgrid.nc', decode_times=False)
ds4 = xr.open_dataset('modis/PAR_Climatology_Apr_remapgrid.nc', decode_times=False)
ds5 = xr.open_dataset('modis/PAR_Climatology_May_remapgrid.nc', decode_times=False)
ds6 = xr.open_dataset('modis/PAR_Climatology_Jun_remapgrid.nc', decode_times=False)
```

```python
Chla_clim = np.concatenate([ds1.chlor_a,ds2.chlor_a,ds3.chlor_a,ds4.chlor_a,
                            ds5.chlor_a,ds6.chlor_a,ds7.chlor_a,ds8.chlor_a,
                            ds9.chlor_a,ds10.chlor_a,ds11.chlor_a,ds12.chlor_a], axis =0)
```

```python
chla_min = np.nanmin(Chla_clim[:,:,:], axis=0)
chla_max = np.nanmax(Chla_clim[:,:,:], axis=0)
chla_dynrng = chla_max-chla_min
```

```python
ds = xr.open_dataset('/glade/work/gabyn/MARBL_spectra/remapgrid.out1.nc')
chla_model = np.nansum(np.stack([ds.ppChl, ds.mp1Chl, ds.diazChl, ds.diat1Chl, ds.mp2Chl, ds.diat2Chl, ds.mp3Chl, ds.diat3Chl, ds.mp4Chl]), axis =0)
chla_model[chla_model<=0]=np.nan
```

## the colorbar that I used before 
norm=colors.LogNorm(vmin=0.01, vmax=1),cmap=cm.nipy_spectral

```python
# averaging all of the months 
fig = plt.figure(figsize=(24, 10))
ax = fig.add_subplot(3, 1, 1, projection=ccrs.PlateCarree(central_longitude=210.0))
cf = ax.pcolor(ds.lon, ds1.lat, np.log10(np.nanmean(chla_model[:,0,:,:], axis = 0)+1), 
               vmin =0, vmax = 0.5, cmap='Spectral_r',transform=ccrs.PlateCarree())
land = ax.add_feature(cartopy.feature.NaturalEarthFeature('physical','land','110m', facecolor='gray'))
ax.set_title('Model Chlorophyll')
cbar = fig.colorbar(cf)
cbar.ax.set_title('$mg/m^3$', fontsize = 10)

ax1 = fig.add_subplot(3, 1, 2, projection=ccrs.PlateCarree(central_longitude=210.0))
cf = ax1.pcolor(ds1.lon, ds1.lat, np.log10(np.nanmean(Chla_clim, axis = 0)+1), 
                vmin = 0, vmax =0.5, cmap='Spectral_r',transform=ccrs.PlateCarree())
land = ax1.add_feature(cartopy.feature.NaturalEarthFeature('physical','land','110m', facecolor='gray'))
ax1.set_title('MODIS Chlorophyll')
cbar = fig.colorbar(cf)
cbar.ax.set_title('$mg/m^3$', fontsize = 10)


ax2 = fig.add_subplot(3, 1, 3, projection=ccrs.PlateCarree(central_longitude=210.0))
cf = ax2.pcolor(ds1.lon, ds1.lat, (np.log10(np.nanmean(chla_model[:,0,:,:], axis = 0)+1)-np.log10(np.nanmean(Chla_clim, axis = 0)+1)),
               vmin =-0.2, vmax =0.2, cmap='RdBu_r',transform=ccrs.PlateCarree())
land = ax2.add_feature(cartopy.feature.NaturalEarthFeature('physical','land','110m', facecolor='gray'))
ax2.set_title('Difference Chlorophyll')
cbar = fig.colorbar(cf)
cbar.ax.set_title('$mg/m^3$', fontsize = 10)
```

```python
# averaging all of the months 
fig = plt.figure(figsize=(24, 10))
ax = fig.add_subplot(3, 1, 1, projection=ccrs.PlateCarree(central_longitude=210.0))
cf = ax.pcolor(ds.lon, ds1.lat, np.log10(np.nanmean(chla_model[:,0,:,:], axis = 0)+1), 
              norm=colors.LogNorm(vmin=0.01, vmax=1),cmap='Spectral_r',transform=ccrs.PlateCarree())
land = ax.add_feature(cartopy.feature.NaturalEarthFeature('physical','land','110m', facecolor='gray'))
ax.set_title('Model Chlorophyll')
cbar = fig.colorbar(cf)
cbar.ax.set_title('$mg/m^3$', fontsize = 10)

ax1 = fig.add_subplot(3, 1, 2, projection=ccrs.PlateCarree(central_longitude=210.0))
cf = ax1.pcolor(ds1.lon, ds1.lat, np.log10(np.nanmean(Chla_clim, axis = 0)+1), 
                norm=colors.LogNorm(vmin=0.01, vmax=1),cmap='Spectral_r',transform=ccrs.PlateCarree())
land = ax1.add_feature(cartopy.feature.NaturalEarthFeature('physical','land','110m', facecolor='gray'))
ax1.set_title('MODIS Chlorophyll')
cbar = fig.colorbar(cf)
cbar.ax.set_title('$mg/m^3$', fontsize = 10)


ax2 = fig.add_subplot(3, 1, 3, projection=ccrs.PlateCarree(central_longitude=210.0))
cf = ax2.pcolor(ds1.lon, ds1.lat, (np.log10(np.nanmean(chla_model[:,0,:,:], axis = 0)+1)-np.log10(np.nanmean(Chla_clim, axis = 0)+1)),
               vmin =-0.2, vmax =0.2, cmap='RdBu_r',transform=ccrs.PlateCarree())
land = ax2.add_feature(cartopy.feature.NaturalEarthFeature('physical','land','110m', facecolor='gray'))
ax2.set_title('Difference Chlorophyll')
cbar = fig.colorbar(cf)
cbar.ax.set_title('$mg/m^3$', fontsize = 10)
```

```python
# averaging all of the months 
fig = plt.figure(figsize=(24, 10))
ax = fig.add_subplot(3, 1, 1, projection=ccrs.PlateCarree(central_longitude=210.0))
cf = ax.pcolor(ds.lon, ds1.lat, np.log10(np.nanmean(chla_model[:,0,:,:], axis = 0)+1), 
               vmin =0, vmax = 0.5, cmap='Spectral_r',transform=ccrs.PlateCarree())
land = ax.add_feature(cartopy.feature.NaturalEarthFeature('physical','land','110m', facecolor='gray'))
ax.set_title('Model Chlorophyll')
cbar = fig.colorbar(cf)
cbar.ax.set_title('$mg/m^3$', fontsize = 10)

ax1 = fig.add_subplot(3, 1, 2, projection=ccrs.PlateCarree(central_longitude=210.0))
cf = ax1.pcolor(ds1.lon, ds1.lat, np.log10(np.nanmean(Chla_clim, axis = 0)+1), 
                vmin = 0, vmax =0.5, cmap='Spectral_r',transform=ccrs.PlateCarree())
land = ax1.add_feature(cartopy.feature.NaturalEarthFeature('physical','land','110m', facecolor='gray'))
ax1.set_title('MODIS Chlorophyll')
cbar = fig.colorbar(cf)
cbar.ax.set_title('$mg/m^3$', fontsize = 10)

chla_diff = (np.nanmean(chla_model[:,0,:,:], axis = 0)-(np.nanmean(Chla_clim, axis = 0)))/(np.nanmean(Chla_clim, axis = 0))
ax2 = fig.add_subplot(3, 1, 3, projection=ccrs.PlateCarree(central_longitude=210.0))
cf = ax2.pcolor(ds1.lon, ds1.lat, np.log10(chla_diff+1),
               vmin =-1, vmax =1, cmap='RdBu_r',transform=ccrs.PlateCarree())
land = ax2.add_feature(cartopy.feature.NaturalEarthFeature('physical','land','110m', facecolor='gray'))
ax2.set_title('Chl model - MODIS')
cbar = fig.colorbar(cf)
cbar.ax.set_title('$mg/m^3$', fontsize = 10)
```

# Now comparing just with the Arcitc Ocean 

```python
# adding the ice 
ice = np.nanmean(ds.ECOSYS_IFRAC, axis=0)
ice[ice<0.82]=np.nan

# making values less than zero into nan
chla = np.nanmean(chla_model[:,1,:,:], axis=0)
chla[chla<=0]=np.nan

fig = plt.figure(figsize=(18,8))
m = Basemap(projection='npstere',round=True, boundinglat=65, lon_0=180+(-100+-30)/2.,resolution='c')
x, y = m(*np.meshgrid(ds.lon,ds.lat))

# Model
ax = fig.add_subplot(131) 
ax.set_title("Chla model")
#chla, lon = add_cyclic_point(np.log10(np.nanmean(chla_model[:,0,:,:], axis = 0)+1), coord=ds.lon, axis=1)
cf = m.pcolor(x,y,np.log10(chla+1), norm=colors.LogNorm(vmin=0.01, vmax=1),
              cmap='Spectral_r')
ax.contourf(x,y,ice,colors='w', levels=1)
m.drawcoastlines(linewidth=0.5, linestyle='solid', color='gray', antialiased=1, ax=None, zorder=None)
m.fillcontinents(color='0.8', lake_color=None, ax=None, zorder=None, alpha=None)
cbar = plt.colorbar(cf, shrink=0.5)
cbar.ax.set_title('$mg Chl/m^3$', fontsize = 10)


# Observations 
ax1 = fig.add_subplot(132) 
ax1.set_title("Chla MODIS")
#chla, lon = add_cyclic_point(np.log10(np.nanmean(Chla_clim, axis = 0)+1), coord=ds1.lon, axis=1)
cf = m.pcolor(x,y, np.log10(np.nanmean(Chla_clim, axis=0)+1), norm=colors.LogNorm(vmin=0.01, vmax=1),
              cmap='Spectral_r')
m.drawcoastlines(linewidth=0.5, linestyle='solid', color='gray', antialiased=1, ax=None, zorder=None)
m.fillcontinents(color='0.8', lake_color=None, ax=None, zorder=None, alpha=None)
cbar = plt.colorbar(cf, shrink=0.5)
cbar.ax.set_title('$mg Chl/m^3$', fontsize = 10)


chla_diff = (np.nanmean(chla_model[:,0,:,:], axis = 0)-(np.nanmean(Chla_clim, axis = 0)))/(np.nanmean(Chla_clim, axis = 0))
ax2 = fig.add_subplot(133) 
ax2.set_title("Chla model - MODIS")
#chla, lon = add_cyclic_point(np.log10(np.nanmean(Chla_clim, axis = 0)+1), coord=ds1.lon, axis=1)
cf = m.pcolor(x,y, np.log10(chla_diff+1),
              cmap='RdBu_r', vmin=-1, vmax =1)
m.drawcoastlines(linewidth=0.5, linestyle='solid', color='gray', antialiased=1, ax=None, zorder=None)
m.fillcontinents(color='0.8', lake_color=None, ax=None, zorder=None, alpha=None)
cbar = plt.colorbar(cf, shrink=0.5)

m.plot
plt.show()

fig.savefig('Chl_ao_mod-obs_obs.png')
```
