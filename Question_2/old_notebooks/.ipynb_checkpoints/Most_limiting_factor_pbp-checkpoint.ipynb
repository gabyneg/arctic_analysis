## Calcualte the temperature limiation term! 
$$r = r_o e^{\frac{-E_a(T_o -T)}{k T_o T}}$$
$$\frac{\mu}{\mu_o} = e^{\frac{-E_a(T_o -T)}{k T_o T}}$$ 

to simplify however, we will create $\alpha = \frac{E_a}{k}$ and $\gamma_T = 
\frac{\mu}{\mu_o}$ so that: 
$$\gamma_T = e^{ \frac{-\alpha(T_o - T)}{T_o T}}$$

$T_o = 298.15 ^oK$
* it is close to the temperature at which many physiological experiments are run, and it is the mean temperature of the subtropical gyres, which represent a large proportion of the worldâ€™s open oceans.

$k = 8.617e^{-5} eVK^{-1}$ 

$E_a = 0.317$ phytoplankton (Kremer et al., 2017).

$E_a = 0.42$ picoplankton 
* Multiple studies have shown that picoplankton have a higher temperature sensitivity compared to phytoplankton of larger sizes (Chen et al., 2014; Stawiarski et al., 2016).

$E_a = 0.65$ zooplankton 
* MTE predicts that heterotrophic organisms will have a greater temperature dependence than autotrophs, as Rubisco carboxylation (rate limiting for photosynthesis) has a lower activation energy (Ea) than ATP synthesis (Allen et al., 2005; Lopez-Urrutia et al., 2006).

and we will have to convert all the surface temperatures from $^oC$ to $K$ and the temperature on the equation and then solve for every gorup... 
$$0Â°C + 273.15 = 273.15K$$


```python
# convert celcius to kelvin 
temp_K = ds.TEMP + 273.15

# constants 
ð‘˜=8.617e-5
ð¸ð‘Ž_phyto = 0.317 
ð¸ð‘Ž_pico  = 0.42
ð¸ð‘Ž_zoo   = 0.65
ð‘‡ð‘œ = 298.15

# create simpler variables. 
ð›¼_phyto = ð¸ð‘Ž_phyto/ð‘˜
ð›¼_pico = ð¸ð‘Ž_pico/ð‘˜
ð›¼_zoo = ð¸ð‘Ž_zoo/ð‘˜

# calculate temperature limitation term
ð›¾ð‘‡_phyto = np.exp((-ð›¼_phyto*(ð‘‡ð‘œ-temp_K))/(ð‘‡ð‘œ*temp_K))
ð›¾ð‘‡_pico = np.exp((-ð›¼_pico*(ð‘‡ð‘œ-temp_K))/(ð‘‡ð‘œ*temp_K))
ð›¾ð‘‡_zoo = np.exp((-ð›¼_zoo*(ð‘‡ð‘œ-temp_K))/(ð‘‡ð‘œ*temp_K))
```

## Find the most limiting factor at each point. 
I want to make a loop that finds the minimum of the limitation terms at every lat and lon and month, and gives an indece of which one it is, then makes that a number from 1-5 where each number will be a different color. 

```python
# diatom 1 
diat1_most_lim_surf = np.nanmin(diat1_surf, axis = 0)
diat1_index = np.zeros((12,180,360)) 
for j in range(12):
    for k in range(180):
        for l in range(360):
            if np.isnan(diat1_most_lim_surf[j,k,l]) == True:
                diat1_index[j,k,l] = np.nan
            if (diat1_most_lim_surf[j,k,l]>0.9) == True:
                diat1_index[j,k,l] = np.nan
                diat1_most_lim_surf[j,k,l] = np.nan
            else:
                diat1_index[j,k,l] = np.argmin(diat1_surf[:,j, k, l])

# diatom 2
diat2_most_lim_surf = np.nanmin(diat2_surf, axis = 0)
diat2_index = np.zeros((12,180,360)) 
for j in range(12):
    for k in range(180):
        for l in range(360):
            if np.isnan(diat2_most_lim_surf[j,k,l]) == True:
                diat2_index[j,k,l] = np.nan
            if (diat2_most_lim_surf[j,k,l]>0.9) == True:
                diat2_index[j,k,l] = np.nan
                diat2_most_lim_surf[j,k,l] = np.nan
            else:
                diat2_index[j,k,l] = np.argmin(diat2_surf[:,j, k, l])

# diatom 3
diat3_most_lim_surf = np.nanmin(diat3_surf, axis = 0)
diat3_index = np.zeros((12,180,360)) 
for j in range(12):
    for k in range(180):
        for l in range(360):
            if np.isnan(diat3_most_lim_surf[j,k,l]) == True:
                diat3_index[j,k,l] = np.nan
            if (diat3_most_lim_surf[j,k,l]>0.9) == True:
                diat3_index[j,k,l] = np.nan
                diat3_most_lim_surf[j,k,l] = np.nan
            
            else:
                diat3_index[j,k,l] = np.argmin(diat3_surf[:,j, k, l])


# mp1
mp1_most_lim_surf = np.nanmin(mp1_surf, axis = 0)
mp1_index = np.zeros((12,180,360)) 
for j in range(12):
    for k in range(180):
        for l in range(360):
            if np.isnan(mp1_most_lim_surf[j,k,l]) == True:
                mp1_index[j,k,l] = np.nan
            if (mp1_most_lim_surf[j,k,l]>0.9) == True:
                mp1_index[j,k,l] = np.nan
                mp1_most_lim_surf[j,k,l] = np.nan
            else:
                mp1_index[j,k,l] = np.argmin(mp1_surf[:,j, k, l])

# mp2
mp2_most_lim_surf = np.nanmin(mp2_surf, axis = 0)
mp2_index = np.zeros((12,180,360)) 
for j in range(12):
    for k in range(180):
        for l in range(360):
            if np.isnan(mp2_most_lim_surf[j,k,l]) == True:
                mp2_index[j,k,l] = np.nan
            if (mp2_most_lim_surf[j,k,l]>0.9) == True:
                mp2_index[j,k,l] = np.nan
                mp2_most_lim_surf[j,k,l] = np.nan
            else:
                mp2_index[j,k,l] = np.argmin(mp2_surf[:,j, k, l])
            
# mp3
mp3_most_lim_surf = np.nanmin(mp3_surf, axis = 0)
mp3_index = np.zeros((12,180,360)) 
for j in range(12):
    for k in range(180):
        for l in range(360):
            if np.isnan(mp3_most_lim_surf[j,k,l]) == True:
                mp3_index[j,k,l] = np.nan
            if (mp3_most_lim_surf[j,k,l]>0.9) == True:
                mp3_index[j,k,l] = np.nan
                mp3_most_lim_surf[j,k,l] = np.nan
            else:
                mp3_index[j,k,l] = np.argmin(mp3_surf[:,j, k, l])
            
# mp4
mp4_most_lim_surf = np.nanmin(mp4_surf, axis = 0)
mp4_index = np.zeros((12,180,360)) 
for j in range(12):
    for k in range(180):
        for l in range(360):
            if np.isnan(mp4_most_lim_surf[j,k,l]) == True:
                mp4_index[j,k,l] = np.nan
            if (mp4_most_lim_surf[j,k,l]>0.9) == True:
                mp4_index[j,k,l] = np.nan
                mp4_most_lim_surf[j,k,l] = np.nan
            else:
                mp4_index[j,k,l] = np.argmin(mp4_surf[:,j, k, l])

# pp
pp_most_lim_surf = np.nanmin(pp_surf, axis = 0)
pp_index = np.zeros((12,180,360)) 
for j in range(12):
    for k in range(180):
        for l in range(360):
            if np.isnan(pp_most_lim_surf[j,k,l]) == True:
                pp_index[j,k,l] = np.nan
            if (pp_most_lim_surf[j,k,l]>0.9) == True:
                pp_index[j,k,l] = np.nan
                pp_most_lim_surf[j,k,l] = np.nan
            else:
                pp_index[j,k,l] = np.argmin(pp_surf[:,j, k, l])
            
# diaz
diaz_most_lim_surf = np.nanmin(diaz_surf, axis = 0)
diaz_index = np.zeros((12,180,360)) 
for j in range(12):
    for k in range(180):
        for l in range(360):
            if np.isnan(diaz_most_lim_surf[j,k,l]) == True:
                diaz_index[j,k,l] = np.nan
            if (diaz_most_lim_surf[j,k,l]>0.9) == True:
                diaz_index[j,k,l] = np.nan
                diaz_most_lim_surf[j,k,l] = np.nan
            else:
                diaz_index[j,k,l] = np.argmin(diaz_surf[:,j, k, l])
```
