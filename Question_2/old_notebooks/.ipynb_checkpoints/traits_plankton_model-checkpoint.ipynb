# Understanding each plankton size 
### plankton descriptors and how they change with volume. 

```python
%matplotlib inline
import warnings
warnings.simplefilter("ignore") # Silence warnings
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
from matplotlib.pyplot import figure
from sklearn.linear_model import LinearRegression
import scipy.io as sio
from numpy import arange,array,ones
from scipy import stats
import seaborn as sns
```

## volume of each plankton. 

```python
# units [um^3]
pp_v    = 0.195428825646897
mp1_v   = 37.6111075368976
diaz_v  = 122.786536059293
diat1_v = 4187.92236603257
mp2_v   = 7238.41739042113
diat2_v = 374617.983936749
mp3_v   = 1393064.17037973
diat3_v = 33510323.645704
mp4_v   = 268101116.324664

phyto_vol = array([pp_v,mp1_v, diaz_v, 
                   diat1_v,mp2_v,diat2_v,
                   mp3_v,diat3_v,mp4_v])
```

### Descriptors

#### $k_{NO_3}$

```python
diaz_kNO3 = 2
pp_kNO3   = 0.0885720275974374
mp1_kNO3  = 0.429135889221455
mp2_kNO3  = 2.07918477665309
mp3_kNO3  = 10.0737538948533
mp4_kNO3  = 48.8078398195214
diat1_kNO3= 0.294068384906009
diat2_kNO3= 1.13220569826432
diat3_kNO3= 4.35915524748407

kNO3_plankton = array([pp_kNO3,mp1_kNO3,diaz_kNO3,
                         diat1_kNO3,mp2_kNO3,diat2_kNO3,
                         mp3_kNO3,diat3_kNO3,mp4_kNO3])
```

```python
# linear regression 
X = np.log10(phyto_vol)
Y = np.log10(kNO3_plankton)

# ploting the figure with the linear regression. 
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
sns.regplot(x=X,y=Y,truncate=True)
m, c, r_value, p_value, std_err = stats.linregress(X, Y)
plt.text(0,0,'$y={:.2f}+{:.2f}x$'.format(c,m),size=14,color="steelblue")

ax.set_xlabel('Volume ($\mu m^3$)',fontsize=16)
ax.set_ylabel('$k_{NO_3}$',fontsize=16)
ax.set_title('phytoplankton $k_{NO_3}$',fontsize=16)

# axes defining where the plankton types are. 
ax = ax.twiny()
ax.set_xlim(ax.get_xlim())
ax.set_xticks(X)
ax.set_xticklabels(['picoplankton','mp1','diaz','diat1','mp2',
                    'diat2','mp3','diat3','mp4'])
plt.show()

fig.savefig('kNO3_vol_phyto.png', dpi=300, bbox_inches='tight');
```
